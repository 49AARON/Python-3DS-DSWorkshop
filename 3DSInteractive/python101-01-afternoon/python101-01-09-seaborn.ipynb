```python
import pandas as pd 
import seaborn as sns
```

## 9.1 Import Flat File CSV

```python
# Read the dataset
df = pd.read_csv("input-data/Pokemon.csv", index_col = 0)
```

```python
df.head()
```

## 9.2 Bar Plot
- A Bar Plot is used to represent a comparison between categories of data. 
- It can be represented either in vertical/horizontal manner.

```python
# Up Size
plt.figure(figsize=(10,6))

# Count Plot
sns.countplot(x='Type 1',data=df)
```

#### Can we Adjust more X-Axis label for better readble

```python
# Up Size
plt.figure(figsize=(10,6))

#Count Plot
sns.countplot(x='Type 1',data=df,palette='rainbow')
plt.xticks(rotation=70)


# ‘rcParams’ method of matplotlib to increase the size of x-axis values and labels of x-axis.
plt.rcParams['xtick.labelsize'] = 15  
plt.rcParams['axes.labelsize'] = 20  
```

## 9.3 Box Plot
- the visual representation statistical five number summary of a given data set i.e. Minimum, First Quartile, Median, Third Quartile and Maximum.
- used in exploratory data analysis. Through this, we can represent the shape of the distribution of data through it.

<img src="images/quartile.png">

```python
# Up Size
plt.figure(figsize=(10,6))


#Boxplot
sns.boxplot(data=df)
plt.xticks(rotation=90)
```

```python
# Try drop 3 columns of distinguished
df_drop = df.drop(columns=['Total','Generation','Legendary'])

# Up Size
plt.figure(figsize=(10,6))
sns.boxplot(data=df_drop)
```

## 9.4 Scatter Plot
- Compare the relationship of two columns and helps us to find potential relationships between values and they can help in detecting outliers in the datasets. 
- They can simplify the large dataset’s representation easily.

```python
# Up Size
plt.figure(figsize=(10,6))
sns.lmplot(x="Sp. Atk", y="Sp. Def", data=df)
```

```python
# Up Size
plt.figure(figsize=(10,6))
sns.lmplot(x="Sp. Atk",y="Sp. Def", data=df, hue='Type 1', fit_reg=False)
```

## 9.5 Violin Plot
- help in representing a fantastically huge amount of information effectively in the small amount of space.

```python
# Setting the theme
sns.set_style('darkgrid')
plt.figure(figsize=(10,6))
 
# Violin Plot
vio_plot = sns.violinplot(x='Type 1', y='Attack', data=df)

plt.xticks(rotation=30)
plt.rcParams['xtick.labelsize'] = 15
plt.rcParams["axes.labelsize"] = 15
```

## 9.6 Swarm Plot
- a better representation of the distribution of values. It can be built on its own but is also a good complement to a box or violin plot.

```python
plt.figure(figsize=(10,6))
plt.xticks(rotation=30)
 
sns.swarmplot(x='Type 1', y='Attack', data=df, dodge=False) 
 
plt.rcParams['xtick.labelsize'] = 15
plt.rcParams["axes.labelsize"] = 15
```

## 9.7 Overlaying Swarm and Violin Plots
- overlaying help us to analyse in a much more efficient manner.
- helps in visualizing the distribution of data and the probability density.

```python
#Setting the figure with matplotlib
plt.figure(figsize=(10,6))
plt.xticks(rotation=30)
plt.rcParams["axes.labelsize"] = 15
 
#Creating the desired plot
sns.violinplot(x='Type 1',y='Attack',data=df,
              inner=None #removes the inner bars inside the violins 
              )
 
sns.swarmplot(x='Type 1',y='Attack',data=df)
        
#Title for the plot
plt.title('Attacks by various types')
```

#### We can make distributed of plot more clear by using "marker" Black color

```python
#Setting the figure with matplotlib
plt.figure(figsize=(10,6))
plt.xticks(rotation=30)
plt.rcParams["axes.labelsize"] = 15
 
#Creating the desired plot
sns.violinplot(x='Type 1',y='Attack',data=df,
              inner=None #removes the inner bars inside the violins 
              )
 
sns.swarmplot(x='Type 1',y='Attack',data=df,
             color='k',#for making the points black
             alpha=0.6) #value of alpha will increase the transparency
 
#Title for the plot
plt.title('Attacks by various types')
```

## 9.8 Joint Kernel Density Plot
- helps in finding the probability density function of any dataset. 
- This can help in smoothing the data around values of PDF ( Probability density function ). 

```python
sns.set_style('whitegrid')
 
plt.figure(figsize=(10,6))
plt.rcParams['xtick.labelsize'] = 15
plt.rcParams['ytick.labelsize'] = 15
 
sns.kdeplot(df.Attack,df.Defense)
```

## 9.10 Joint Plot
-  get the liberty to use two plots for representation of the same data which helps in a better analysis.

```python
plt.figure(figsize=(10,6))
sns.jointplot(x="Sp. Atk",y='Sp. Def',data=df,kind='scatter',color='b')
```

## 9.11 Heatmap
- it is the 2-dimensional figure 
- helps in visualizing complex data. We can even use it to represent large datasets.

```python
flights = sns.load_dataset('flights')
```

#### Displaying the flight’s dataset.

```python
flights.head()
```

```python
# Create Pivot Table
flights = flights.pivot('month','year','passengers')
```

```python
flights.head()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(flights,cmap="OrRd")
```

- ‘annot’ argument which is set as ‘True’
- for displaying those values in a decimal format we use the ’fmt’ argument. 
- draw borders between the lines using ‘linewidths’ argument

```python
plt.figure(figsize=(10,6))
sns.heatmap(flights,cmap="coolwarm",annot=True,fmt='d',linewidths=2)
plt.xticks(rotation=70)
```

## 9.12 Clustermap
- ability to standardize data.
- represent datasets through a cluster map which have columns with huge differences making it easier for analysis.

```python
plt.figure(figsize=(15,10))
sns.clustermap(flights,cmap='RdYlBu',linewidths=0.5,annot=True,fmt='d')
```

- standard_scale for standardizing our dataset 
- use ‘0’ for the data across rows will be standardized
- if we pass ‘1’ the column data will be standardized.

```python
sns.clustermap(flights,col_cluster=False,cmap='cool',linewidths=2,standard_scale =1 ,annot=True)
```
