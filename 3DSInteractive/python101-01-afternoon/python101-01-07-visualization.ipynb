## 7.1 Basic plots

```python
import numpy as np
import matplotlib.pyplot as plt
# inline plot (for jupyter)
%matplotlib inline
```

```python
x = np.linspace(0, 10, 50)
sinus = np.sin(x)
plt.plot(x, sinus)
plt.title('My First Plot')
plt.xlabel('this is x!')
plt.ylabel('this is y!')
```

```python
plt.plot(x,sinus, 'go--', linewidth=2, markersize=5)
plt.title('My First Plot')
plt.xlabel('this is x!')
plt.ylabel('this is y!')
```

```python
plt.plot(x, sinus, "o")
plt.show()
```

```python
# Rapid multiplot
cosinus = np.cos(x)
plt.plot(x, sinus, "-b", x, sinus, "ob", x, cosinus, "-r", x, cosinus, "or")
plt.xlabel('this is x!')
plt.ylabel('this is y!')
plt.title('My First Plot')
plt.show()
```

```python
# Step by step
plt.plot(x, sinus, label='sinus', color='blue', linestyle='--', linewidth=2)
plt.plot(x, cosinus, label='cosinus', color='red', linestyle='-', linewidth=2)
plt.xlabel('this is x!')
plt.ylabel('this is y!')
plt.title('My First Plot')
plt.legend()
plt.show()
```

## 7.2 Scatter (2D) plots

```python
import pandas as pd 
```

```python
path = 'input-data/salary_table.csv'
salary = pd.read_csv(path)
df = salary
```

```python
df.head()
```

```python
df.shape
```

### 7.2.1 Simple scatter with colors

```python
colors = {'Bachelor':'r', 'Master':'g', 'Ph.D':'blue'}
```

```python
plt.figure(figsize=(8,5))
plt.scatter(df['experience'], df['salary'], c=df['education'].apply(lambda x: colors[x]), s=100)
plt.xlabel('Year Experiences')
plt.ylabel('Salary')
plt.title('Salary vs Year')
```

### 7.2.2 Scatter plot with colors and symbols

```python
## Define colors / sumbols manually
symbols_manag = dict(Y='*', N='.')
colors_edu = {'Bachelor':'r', 'Master':'g', 'Ph.D':'blue'}
```

```python
## Figure size
plt.figure(figsize=(10,8))

## group by education x management => 6 groups
for values, d in salary.groupby(['education','management']):
    edu, manager = values
    plt.scatter(d['experience'], d['salary'], marker=symbols_manag[manager],color=colors_edu[edu],s=200, label=manager+"/"+edu)
## Set labels
plt.xlabel('Experience') 
plt.ylabel('Salary') 
plt.title('Salary vs Year')
plt.legend(loc=4) # lower right 
plt.show()
```

```python
for values, d in salary.groupby(['education','management']):
    print(values, d)
```
